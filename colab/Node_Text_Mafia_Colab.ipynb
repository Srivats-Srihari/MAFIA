```python
from google.colab import drive
import os

drive.mount('/content/drive')

# Change this to your folder path in Drive
PROJECT_DIR = '/content/drive/MyDrive/node-text-mafia'

if not os.path.isdir(PROJECT_DIR):
    raise FileNotFoundError(f'Folder not found: {PROJECT_DIR}')

print('Using project dir:', PROJECT_DIR)
```

```python
import subprocess

subprocess.run(['npm', 'install'], cwd=PROJECT_DIR, check=True)
```

```python
import os

# Recommended for Colab
os.environ['MAFIA_HEADLESS'] = '1'
os.environ['MAFIA_GUI_HOST'] = '0.0.0.0'
os.environ['MAFIA_GUI_PORT'] = '8787'

# Optional provider config examples:
# os.environ['MAFIA_AI_PROVIDER'] = 'sambanova'
# os.environ['SAMBANOVA_API_KEY'] = '...'
# os.environ['GROQ_API_KEY'] = '...'
# os.environ['OPENROUTER_API_KEY'] = '...'
```

```python
import os
import time
import subprocess
from google.colab.output import eval_js

port = int(os.environ.get('MAFIA_GUI_PORT', '8787'))

proc = subprocess.Popen(
    ['npm', 'run', 'colab:gui'],
    cwd=PROJECT_DIR,
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT,
    text=True
)

time.sleep(4)
url = eval_js(f'google.colab.kernel.proxyPort({port})')
print('Open GUI:', url)
```

```python
# Optional: stream server logs
while True:
    line = proc.stdout.readline()
    if not line:
        break
    print(line, end='')
```
